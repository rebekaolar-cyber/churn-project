# EDA


```python
import pandas as pd
import plotly.express as px

# Load dataset
file_path = "../data/raw/telco_churn.csv"
df = pd.read_csv(file_path)

# Display basic info
print("Shape:", df.shape)
print("\nData types:")
print(df.dtypes)

print("\nMissing values per column:")
print(df.isna().sum())

# Inspect potential problematic columns
cols_to_inspect = ["TotalCharges"]
for col in cols_to_inspect:
    if col in df.columns:
        print(f"\nUnique values sample for {col}:")
        print(df[col].value_counts(dropna=False).head(10))

# Plot distribution of target variable 'Churn'
if "Churn" in df.columns:
    fig = px.histogram(df, x="Churn", title="Distribution of Churn")
    fig.show()
else:
    print("Column 'Churn' not found in dataframe.")
```
